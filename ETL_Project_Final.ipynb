```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
import numpy as np
import getpass
password = getpass.getpass()
import psycopg2
```

```python
csv_file1 = "Resources/worldwide_data.csv"
worldwide_data_df = pd.read_csv(csv_file1)

worldwide_data_df
```

```python
worldwide_data_df.dropna(subset="continent")
worldwide_data_df['continent'].unique()
continent_df = pd.DataFrame({"continent": worldwide_data_df['continent'].unique()})
continent_df = continent_df.reset_index()
continent_df["continent_id"] = continent_df["continent_id"] + 1
continent_df
```

```python
worldwide_data_df['location'].unique()
countries_df = pd.DataFrame({"country": worldwide_data_df['location'].unique()})
countries_df = countries_df.reset_index()
countries_df["country_id"] = countries_df["country_id"] + 1
countries_df = countries_df.rename(columns = {"country_id": "country_id"})
countries_df
```

```python
worldwide_data_df['iso_code'].unique()
iso_df = pd.DataFrame({"iso_code": worldwide_data_df['iso_code'].unique()})
iso_df = iso_df.reset_index()
iso_df = iso_df.rename(columns = {"index": "iso_id"})
iso_df["iso_id"] = iso_df["iso_id"] + 1
iso_df
```

```python
worldwide_data_df = pd.merge(continent_df, worldwide_data_df, how="inner", left_on="continent", right_on="continent")
```

```python
worldwide_data_df = pd.merge(countries_df, worldwide_data_df, how="inner", left_on="country", right_on="location")
```

```python
worldwide_data_df = pd.merge(iso_df, worldwide_data_df, how="inner", left_on="iso_code", right_on="iso_code")
```

```python
del worldwide_data_df["location"], worldwide_data_df["continent"], worldwide_data_df["iso_code"]
```

```python
worldwide_data_df
```

```python
worldwide_data_df = worldwide_data_df.reset_index()
worldwide_data_clean = worldwide_data_df[["index", "continent_id", "country_id", "iso_id", "date", "total_cases", "new_cases", "total_deaths", "new_deaths", "hosp_patients"]]
```

```python
worldwide_clean_renamed = worldwide_data_clean.rename(columns = {"index": "id", "new_cases": "daily_cases", "new_deaths": "daily_deaths"})
worldwide_clean_final = worldwide_clean_renamed.fillna(0)
worldwide_clean_final["id"] = worldwide_clean_final["id"] + 1
worldwide_clean_final
```

```python
rds_connection_string = f'postgres:{password}@localhost:5432/covid_db'
engine = sqlalchemy.create_engine(f'postgresql://{rds_connection_string}')
conn = engine.connect()
```

```python
engine.table_names()
```

```python
continent_df.to_sql(name='continents', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('SELECT * FROM continents', con=engine)
```

```python
iso_df.to_sql(name='iso', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from iso', con=engine)
```

```python
countries_df.to_sql(name='countries', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from countries', con=engine)
```

```python
worldwide_clean_final.to_sql(name='worldwide', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from worldwide', con=engine)
```
